```python
data = pd.read_sql('''SELECT
CustomerID,
MAX(TransactionDate) AS Last_Order,
COUNT(TransactionID) AS Frequency,
SUM("TransactionAmount (INR)") AS Monetary
FROM BankTransactions1
WHERE CustomerID IS NOT NULL
GROUP BY CustomerID''', conn)
data
```

```python
import numpy as np
import datetime
# Check the data types
data.dtypes
data['Last_Order'] = pd.to_datetime(data['Last_Order'])
data.dtypes

# Create snapshot date
snapshot_date = data['Last_Order'].max() + datetime.timedelta(days=1)
print(snapshot_date)
# Aggregate data by each customer
customers = data.groupby(['CustomerID']).agg({
'Last_Order': lambda x: (snapshot_date - x.max()).days, # Use 'last_order' here
'Frequency':'sum' ,
'Monetary': 'sum'})
# Rename columns
customers.rename(columns = {'Last_Order': 'Recency', # Rename 'last_order' to 'Recency'
'Frequency': 'Frequency',
'Monetary': 'Monetary'}, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.histplot(data=customers, x='Recency', kde=True)
ax.set_title('Histogram and Recency')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot RFM distributions
plt.figure(figsize=(12,10))
# Plot distribution of R
plt.subplot(3, 1, 1); sns.distplot(customers['Recency'])
# Plot distribution of F
plt.subplot(3, 1, 2); sns.distplot(customers['Frequency'])
# Plot distribution of M
plt.subplot(3, 1, 3); sns.distplot(customers['Monetary'])
# Show the plot
plt.show()
```

```python
from scipy import stats
customers_data = pd.DataFrame()
customers_data['Recency'] = stats.boxcox(customers['Recency'])[0]
customers_data['Frequency'] = stats.boxcox(customers['Frequency'])[0]
customers_data['Monetary'] = stats.boxcox(customers['Monetary'] + 1e-8)[0]
customers_data.tail()
# Plot RFM distributions
plt.figure(figsize=(12,10))
# Plot distribution of R
plt.subplot(3, 1, 1); sns.distplot(customers_data['Recency'])
# Plot distribution of F
plt.subplot(3, 1, 2); sns.distplot(customers_data['Frequency'])
# Plot distribution of M
plt.subplot(3, 1, 3); sns.distplot(customers_data['Monetary'])
# Show the plot
plt.show()
```

```python
# Import library
from sklearn.preprocessing import StandardScaler
# Initialize the Object
scaler = StandardScaler()
# Fit and Transform The Data
scaler.fit(customers_data)
customers_normalized = scaler.transform(customers_data)
# Assert that it has mean 0 and variance 1
print(customers_normalized.mean(axis = 0).round(2))
print(customers_normalized.std(axis = 0).round(2))
```

```python
from sklearn.cluster import KMeans
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(customers_normalized)
    sse[k] = kmeans.inertia_ # SSE to closest cluster centroid
plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()

```

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12))
visualizer.fit(customers_normalized)
visualizer.show()
```

```python
model = KMeans(n_clusters=4, random_state=42)
model.fit(customers_normalized)
model.labels_.shape
customers["Cluster"] = model.labels_
customers.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'Monetary':['mean', 'count']}).round(2)
f, ax = plt.subplots(figsize=(25, 5))
ax = sns.countplot(x="Cluster", data=customers)
customers.groupby(['Cluster']).count()
```
